```python
import pandas as pd 

# Load data
df = pd.read_csv("superstore.csv", encoding='latin1')

print(df.info())
print(df.head())
print(df.isnull().sum())

# Fix Missing Values
df['Postal Code'] = df['Postal Code'].fillna(0).astype(int)

# Fix dates 
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)

# Delivery days
df['delivery_days'] = (df['Ship Date'] - df['Order Date']).dt.days

# Q1: Region with highest sales
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print("Region Sales:\n", region_sales)

# Q2: Most profitable category
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print("\nCategory Profit:\n", category_profit)

# Q3: Loss-making sub-categories
loss_df = df[df['Profit'] < 0]
loss_subcat = loss_df.groupby('Sub-Category')['Profit'].sum().sort_values()
print("\nLoss Sub-Categories:\n", loss_subcat)

# Q4: Average delivery time
avg_delivery = df['delivery_days'].mean()
print("\nAverage Delivery Days:", avg_delivery)

# Q5 Top 10 products

top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Products:\n", top_products)

df.to_csv("cleaned_superstore.csv", index=False)
```
